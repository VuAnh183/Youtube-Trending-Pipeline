```python
with open(file_path, "r") as f:
    data = json.load(f)

filtered_df = pd.json_normalize(data['items'])

filtered_df
```

***Data features to keep***

- items['id']  
- items['snippet.publishedAt']  
- items['snippet.channelId']  
- items['snippet.title']  
- items['snippet.channelTitle']  
- items['snippet.categoryId']  
- items['statistics.viewCount']  
- items['statistics.likeCount']  
- items['statistics.commentCount']  

```python
columns_to_keep = [
    'id',
    'snippet.publishedAt',
    'snippet.channelId',
    'snippet.title',
    'snippet.channelTitle',
    'snippet.categoryId',
    'statistics.viewCount',
    'statistics.likeCount',
    'statistics.commentCount'
]
```

```python
clean_df = filtered_df[columns_to_keep].copy()

clean_df
```

***Rename all the fields***

```python
clean_df.rename(columns={
    'id': 'video_id',
    'snippet.title': 'title',
    'snippet.publishedAt': 'published_at',
    'snippet.channelId': 'channel_id',
    'snippet.channelTitle': 'channel_title',
    'snippet.categoryId': 'category_id',
    'statistics.viewCount': 'views',
    'statistics.likeCount': 'likes',
    'statistics.commentCount': 'comments'
}, inplace=True)
```

```python
clean_df
```

***Convert types***

```python
clean_df['views'] = clean_df['views'].astype(int)
clean_df['likes'] = clean_df['likes'].astype(int)
clean_df['comments'] = clean_df['comments'].astype(int)
clean_df['published_at'] = pd.to_datetime(clean_df['published_at'])
```
